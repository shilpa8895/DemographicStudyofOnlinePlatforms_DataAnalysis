```python
import os
import pandas as pd

# Set the directory containing your CSV files
directory = '/Users/sinaroman/Data_Repository'

# Create an empty list to hold dataframes
dataframes = []

# Loop through each file in the directory
for filename in os.listdir(directory):
    if filename.endswith('.csv'):
        # Read the CSV file into a dataframe, ignoring the original index column
        filepath = os.path.join(directory, filename)
        df = pd.read_csv(filepath)
        
        # Create the 'Service Name' column
        service_name = filename.replace('_', ' ').replace('.csv', '')
        df['Service Name'] = service_name
        
        # Append the dataframe to the list
        dataframes.append(df)

# Concatenate all dataframes in the list
merged_df = pd.concat(dataframes, ignore_index=True)

# Reset the index to start from 1
merged_df.index = merged_df.index + 1

# Save the merged dataframe to a new CSV file
output_path = '/Users/sinaroman/Stackoverflow_Data.csv'
merged_df.to_csv(output_path, index_label='Index')
```
